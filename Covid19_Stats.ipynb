```python
t1=datetime.now()
t1
```

```python
#geocoding the addresses
import geopy
from geopy.geocoders import ArcGIS
nom=ArcGIS()
from geopy.exc import GeocoderTimedOut

def do_geocode(address):
    try:
        return nom.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)
   
   

```

```python
print(f"Geocoding Started {datetime.now()}")

Covid["Location"] = Covid["Country"].apply(do_geocode)

print(f"Geocode Complete: {datetime.now()-t1}")
```

```python
Covid["latlong"]=Covid["Location"].apply(lambda x:[x.latitude,x.longitude] if x != None else None)
```

```python
import matplotlib.pyplot as plt
import folium
%matplotlib Qt
from folium.plugins import MarkerCluster
```

```python

def color_selector(TotalConfirmed):
    try:
        #price = float(price.replace("$", "").replace(",", ""))
        if TotalConfirmed <= 600:
            return "green"
        elif TotalConfirmed > 600 and TotalConfirmed <= 1000:
            return "orange"
        else:
            return "red"
    except:
        return "black"
```

```python
subset_of_df2 = df2.sample(n=len(df2))
some_map=folium.Map(location=[53.9333, 116.5765], zoom_start=4)
#creating a Marker for each point in df_sample. Each point will get a popup with their zip
for row in subset_of_df2.itertuples():
    some_map.add_child(folium.Marker(location=[row.latlong[0],row.latlong[1]],
           popup=["Country :" + row.Country,"NewConfirmed :" + str(row.NewConfirmed)]))
some_map
some_map.save("covid.html")
```

```python



#Let’s look at those same set of points using MarkerClusters.
from folium.plugins import MarkerCluster
maped=folium.Map(location=[53.9333, 116.5765], zoom_start=13,tiles='Stamen Terrain')
# tooltip = 'Click me!'
mc = MarkerCluster()
for row in subset_of_df2.itertuples():
     mc.add_child(folium.Marker(location=[row.latlong[0],row.latlong[1]],
           popup=["Country :" + row.Country, row.TotalConfirmed],tooltip='<b>Click for More Info.</b>'))
maped.add_child(mc)
maped.save("covid2.html")
```
